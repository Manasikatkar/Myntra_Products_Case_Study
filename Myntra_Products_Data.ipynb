```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('myntra_products_catalog.csv')
df
```

```python
# take first and last five items of given data
df.head()
```

```python
df.tail()
```

```python
#find all null values
df.isnull().sum()
```

```python
# info
df.info()
```

```python
# drop null values of price(inr)
df.dropna(inplace = True, subset = ['Price (INR)'])
df
```

```python
# drop column of productname and description
#data = df

df.drop(['ProductName','Description'], inplace = True, axis = 1)
df
```

```python
# check most price of productbrand
df['ProductBrand'].value_counts()
```

```python
# check the higest price of product
plt.figure(figsize=(7,7))
sns.countplot(data = df, x = df['ProductBrand'], order = df.ProductBrand.value_counts().index[:10], color = 'green')
plt.xlabel('ProductBrand',fontsize=10)
plt.ylabel('counts',fontsize=10)
plt.xticks(rotation = 90)
plt.show()
```
